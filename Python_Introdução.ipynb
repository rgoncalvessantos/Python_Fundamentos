### `INT()`

O `int()` (inteiro) é um tipo composto por caracteres numéricos (algarismos), utilizado para representar um número que pode ser escrito sem um componente decimal, podendo ter ou não sinal, isto é, ser positivo ou negativo. Os valores do tipo inteiro podem ter um tamanho ilimitado

#### Exemplo de `INT`

````python
numero1 = float(5)
numero2 =-45
numero3 = 1998439047137419471947198347198748134937591375981375389571309510397589237492871095894279817897198

print(numero1, numero2, numero3)
````

### `FLOAT()`

O `float()` (flutuante) também é composto por caracteres numéricos, mas é utilizado para representar números racionais (números que podem ser escritos como frações),
também conhecidos como ponto flutuante. Assim como os inteiros, podem ter sinal ou não e possuem um tamanho ilimitado. As casas decimais são separadas da porção
inteira por um ponto (.)

#### Exemplo de `FLOAT()`

````python
x = 1289.1298312371237981

print(x)
````

#### Exemplo `str()`


````python
corsa = 'Meu carro é um corsa'
fusca = "Meu carro é um fusca"
gol = """Hoje eu comi bolo
vim para o senai
e estou estudando"""

print(gol)
````

## Operadores Aritméticos

- Simbolos
- Como utilizar

O Python possuí sete operadores aritméticos que são utilizados para realizar operações matemáticas como adição, subtração e multiplicação. São eles:

| Símbolo | Operação | Descrição | Utilização |
| --- | --- | --- | --- |
| + | Adição | Soma os dois valores | x + y |
| - | Subtração | Subtrai o valor da esquerda para a direita | x - y |
| * | Multiplicação | Multiplica os dois valores | x * y |
| / | Divisão | Divide o valor da esquerda pelo valor da direita. O resultado sempre retorna como Float | x / y |
| % | Módulo | Resto da divisão do valor da esquerda pelo da direita. O resultado sempre retorna Int | x % y |
| // | Divisão Inteira | Divisão inteira do valor da esquerda pelo valor da direita. O resultado sempre é Int | x // y |
| ** | Exponenciação | Realiza a potencialização do valor da esquerda, elevando ele ao valor da direita | x ** y |


Cada operador tem sua utilização dentro do necessário, no Python quado for calcular algum valor ele usa a convenção tradicional da matemática muito conhecida como **PEMDAS**

| Letra | Significado |
| --- | --- |
| P | Parênteses |
| E | Exponenciação |
| M | Multiplicação |
| D | Divisão |
| A | Adição |
| S | Subtração |

````python
soma = 5 + 40
print(soma)

subtracao = 4 - 5
print(subtracao)

multiplicacao = 1000 * 2
print(multiplicacao)

divisao = 7 / 4
print(divisao)

modulo = 5 % 2
print(modulo)

divisao_inteira = 7 // 4
print(divisao_inteira)

elevado = 5 ** 2
print(elevado)
````

````python
saldo = 358.34
deposito = 345.35

total = saldo * deposito

print(f'Seu total atual é R${total:.2f}')
````

## Entrada de Informação e Comentários



### Entrada de informação

No Python conseguimos colocar informações em nossa variáveis sem que seja nescessário a utilização daquela informação ja pronta.

Com a utilização da função `input()` conseguimos fazer com que nosso usuário possa entrar com a informação e assim podemos tratar ela como acharmos necessário

#### Exemplo de `INPUT()`

````python
nome = input('Digite seu nome: ')

print(f'Seja Bem-Vindo {nome}')
````

````python
n1 = int(input('Digite o Primeiro numero: '))
n2 = int(input('Digite o Segundo numero: '))

soma = n1 + n2

print(f'A soma de {n1} + {n2} = {soma}')
type(n1)
````

Crie um programa que tire a média entre dois números que o usuário vai inserir

````python
"""Pedi para o usuario adicionar dois numeros
depois tirei
mostrei para ele o resultado da média
"""
n1 = int(input('Digite a primeira nota: '))
n2 = int(input('Digite a segunda nota: '))

media = int((n1 + n2) / 2)

print(f'{media}')
````

### Comentários

Comentários são trechos de códigos que são ignorados pelo interpretador da linguagem. O uso de comentários auxilia na documentação do programa, pois fornece algumas informações e/ou orientações sobre o seu funcionamento.

Em Python os comentários, são identificados por um sinal de cerquilha (#), e qualquer conteúdo após este símbolo é ignorado pelo interpretador

#### Exemplo de Comentário

### Desafio

Crie um programa que leia dois números inteiros que o usario vai inserir e traga a média desses dois números

````python
# Pedindo para o usuario adicionar um número para tirar média#

num1 = int(input('Digite um numero '))
num2 = int(input('Digite um numero '))

media = (num1 + num2) / 2

print(media)
````
